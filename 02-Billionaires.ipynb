## 3) Who are the top 10 richest billionaires? Use the `networthusbillion` column.

```python
len_sorted = df.sort_values('networthusbillion', ascending=False)
len_sorted_top_ten = len_sorted.head(10)
len_sorted_top_ten
```

## 4a) How many male billionaires are there compared to the number of female billionares?

> **TIP:** This is not a `.groupby`

```python
count_male = df.gender.value_counts().get('male')
count_female = df.gender.value_counts().get('female')

f'There are {count_male} male billionaires and {count_female} female billionaires.'
```

## 4b) What percent is that?

> **TIP:** Doing percentages with the answer to `4a` is so common that there's an extra parameter to automatically calculate it. Look at the documentation!

```python
male_percentage = count_male / (count_male + count_female) * 100
f'{male_percentage: .2f} of billionaires are male.'
```

## 4c) Is the average wealth different for a female billionaire as compared to a male billionaire?

> **TIP:** This one *does* use `.groupby`. You can pick what 'average' means in this context.

```python
male_list = df[df['gender'] == 'male'].copy()
avg_male_wealth = male_list['networthusbillion'].mean()

female_list = df[df['gender'] == 'female'].copy()
avg_female_wealth = female_list['networthusbillion'].mean()
f'The average net worth of male billionaires is {avg_male_wealth: .2f} billion USD, while the average net worth of female billionaires is {avg_female_wealth: .2f} billion USD.'
```

## 5a) What is the most common source/type of wealth?

```python
wealth_types = df.typeofwealth.value_counts()
wealth_types.sort_values(ascending=False).head(1)
```

## 5b) Is the most common source/type of wealth different between men and women?

> **TIP:** You know how to `groupby` and you know how to count how many times a value is in a column. Can you put them together??? You could also do this as two separate cells where you calculate it separately for men and women.
>
> **TIP:** Use percentages for this, it makes it a lot more readable.

```python
wealth_types_male = male_list.typeofwealth.value_counts()
max_wealth_type_male = wealth_types_male.idxmax()

wealth_types_female = female_list.typeofwealth.value_counts()
max_wealth_type_female = wealth_types_female.idxmax()

f'The most common type of wealth among male billionaires is {max_wealth_type_male} and among female billionaires is {max_wealth_type_female}.'
```

## 7) How much money do these top 5 billionaires have in total?

> **TIP:** for multiple conditions, you can't use `or` and `and`. In pandas, `|` means *or* and `&` means *and*.
>
> **TIP:** Checking multiple conditions can be done two ways.
> - `(df.colname == 'A') | (df.colname == 'B')`
> - `df.colname.isin(['A', 'B'])`

```python
richest_sorted = df.sort_values('networthusbillion', ascending=False)
richest_sorted_top_five = len_sorted.head(5)
richest_five_total_wealth = richest_sorted_top_five['networthusbillion'].sum()
f'The five richest billionaires have a combined networth of {richest_five_total_wealth} billion USD.'
```

## 8) What are the top 10 countries with the most money held by billionaires?

I am **not** asking which country has the most billionaires - this is **total amount of money per country.**

> **TIP:** Think about it in steps - "I want them organized by country," "I want their net worth," "I want to add it all up," and "I want 10 of them." Just chain it all together.

```python
total_wealth = df.groupby('countrycode')['networthusbillion'].sum()
total_wealth_sorted = total_wealth.sort_values(ascending=False).head(10)

total_wealth_sorted
```
